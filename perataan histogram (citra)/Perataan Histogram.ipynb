# Perataan Histogram

```python
#Import library cv2, numpy, matplotlib
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#import gambar
gambar = cv2.imread("gambar.jpg")

#menampilkan gambar
plt.imshow(gambar)
```

```python
#karena gambar di atas dalam format warna BGR, 
#maka kita ubah dahulu ke format warna RGB
gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB)

#menampilkan gambar
plt.imshow(gambar)
```

```python
#mengimport dan menampilkan gambar dengan grayscale
#untuk menampilkan grayscale, kita tambahkan cmap = "gray" dari library matplotlib
grayscale_gambar = cv2.imread("gambar.jpg", 0)

plt.imshow(grayscale_gambar, cmap="gray")
```

```python
#melakukan perataan histogram
eq_grayscale_gambar = cv2.equalizeHist(grayscale_gambar)

plt.imshow(eq_grayscale_gambar, cmap="gray")
```

```python
nilai_hist = cv2.calcHist([grayscale_gambar], channels=[0], mask=None, histSize=[256], ranges=[0,256])
nilai_hist.shape
plt.plot(nilai_hist)
```

```python
nilai_hist = cv2.calcHist([eq_grayscale_gambar], channels=[0], mask=None, histSize=[256], ranges=[0,256])

plt.plot(nilai_hist)
```

# Melakukan Perataan Histogram dengan memanfaatkan Value pada HSV

# Sekarang kita bandingkan Grafik Histogram aslinya dengan Grafik Histogram yang sudah dilakukan proses perataan histogram

```python
#mengubah format warna dari RGB ke HSV
hsv_gambar = cv2.cvtColor(gambar, cv2.COLOR_RGB2HSV)

#mengambil nilai dari Value dari HSV
#(Hue, Saturation, Value)
hsv_gambar[:,:,2]

#menghitung nilai histogram dari gambar di atas
#channels=[2] yang berarti kita mencari nilai dari Value
nilai_hist = cv2.calcHist([hsv_gambar], channels=[2], mask=None, histSize=[256], ranges=[0,256])

plt.plot(nilai_hist)
```

```python
#melakukan perataan histogram kepada Value
hsv_gambar[:,:,2] = cv2.equalizeHist(hsv_gambar[:,:,2])

nilai_hist_eq = cv2.calcHist([hsv_gambar], channels=[2], mask=None, histSize=[256], ranges=[0,256])

plt.plot(nilai_hist_eq)
```

# Sekarang kita bandingkan gambar aslinya dengan gambar yang sudah dilakukan proses perataan histogram

```python
#menampilkan gambar asli
plt.imshow(gambar)
```

```python
#merubah kembali format warna dari
#format HSV ke RGB
hasil = cv2.cvtColor(hsv_gambar, cv2.COLOR_HSV2RGB)

#dan inilah gambar setelah dilakukan
#proses perataan histogram
plt.imshow(hasil)
```
